## Setting Up for Inference

```python
from fastai.vision.all import *
```

```python
version = 2
img_path = "../deployment"
model_path = "../models/pasta_shape_recognizer"
```

```python
model = load_learner(f"{model_path}_v{version}.pkl")
```

## Inference 1

**Target:** sphagetti

```python
img = Image.open(f'{img_path}/unknown00.png')
img.to_thumb(128,128)
```

```python
model.predict(f'{img_path}/unknown00.png')
```
